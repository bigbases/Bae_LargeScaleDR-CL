```python
import pandas as pd
data = pd.read_pickle("/root/workspace/AMI/ami/ireland_power_1h.bin")
data = data.set_index("time")
df = data.iloc[:,-4710:]
df
```

```python
import numpy as np

# Select 40 random columns
random_columns_40 = np.random.choice(df.columns, size=40, replace=False)
df_40 = df[random_columns_40]

# Select 400 random columns
random_columns_400 = np.random.choice(df.columns, size=400, replace=False)
df_400 = df[random_columns_400]

# Resample to daily data
df_daily = df.resample("D").sum()
df_40_daily = df_40.resample("D").sum()
df_400_daily = df_400.resample("D").sum()
```

```python
import matplotlib.pyplot as plt

# Daily Data
y_min, y_max = 0, 1300

plt.figure(figsize=(8, 14))

plt.subplot(3, 1, 1)
plt.plot(df_40_daily, linewidth=0.5)
plt.ylabel("Electricity Consumption (kWh)")
plt.ylim(y_min, y_max)

plt.subplot(3, 1, 2)
plt.plot(df_400_daily, linewidth=0.5)
plt.ylabel("Electricity Consumption (kWh)")
plt.ylim(y_min, y_max)

plt.subplot(3, 1, 3)
plt.plot(df_daily, linewidth=0.5)
plt.ylabel("Electricity Consumption (kWh)")
plt.ylim(y_min, y_max)

plt.tight_layout()
plt.show()
```
